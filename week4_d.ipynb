```python
import numpy as np
import matplotlib.pyplot as plt
A=np.array([[7,2], [-7,5]])
b=np.array([-5,12])
x=np.linalg.solve(A,b)

x1,y1=x
print(x1,y1)
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

a1=A[:,0]
b1=A[:,1]
c1=-b

for c1,c2,c3 in zip(A[:,0],A[:,1],b):
  x=np.linspace(-7,7,100)
  y=(c3-c1*x)/c2
  ax.plot(x,y,color="black")

ax.plot(x1,y1,'ro')
ax.axis([-7,7,-7,7])
ax.set_xticks(range(-7,7))
ax.set_yticks(range(-7,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable='box')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread("/content/drive/MyDrive/linearmath.jpg")
height,width =img.shape[:2]
m=np.array([[1,0,100,],[0,0,1]],dtype=float)
dst=np.zeros(img.shape, dtype=np.uint8)
for y in range(height-1):
  for x in range(width-1):
    p=np.array([x,y,1])
    p_=np.dot(m,p)
    x_,y_=p_[:2]
    x_=int(x_)
    y_=int(y_)
    if x_>0 and x_<width and y_ >0 and y_<height :
      dst[y_,x_]=img[y,x]
result=cv2.hconcat([img,dst])
cv2_imshow(result)
cv2.waitKey(0)
```
